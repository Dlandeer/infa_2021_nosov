```python
import numpy as np
import matplotlib.pyplot as plt
x=[]
y=[]
with open("mutant.txt","r") as inp:
    for line in inp:
        if line[0]=="C":
            continue
        x.append(float(line.split()[0]))
        y.append(float(line.split()[1]))
sp=plt.subplot(121)
sp.scatter(x,y)
plt.title('Mutant')
plt.xlabel('Concentration,nM')
plt.ylabel('Response')
x1=[]
y1=[]
with open("wild_type.txt","r") as inp:
    for line in inp:
        if line[0]=="C":
            continue
        x1.append(float(line.split()[0]))
        y1.append(float(line.split()[1]))
sp=plt.subplot(122)
sp.scatter(x1,y1)
plt.title('Wild type')
plt.xlabel('Concentration,nM')
plt.ylabel('Response')
plt.show()
```

```python
import numpy as np
def corr(x,y):
    n=len(x)
    c=np.sum((x-np.mean(x))*(y-np.mean(y)))/np.sqrt(np.sum((x-np.mean(x))**2)*np.sum((y-np.mean(y))**2))
    return c
x1 = [1, 2, 3, 4, 5]
y1 = [0.99, 0.49, 0.35, 0.253, 0.18]
c1=corr(x1,y1)
x2 = np.random.rand(100)
y2 = np.random.rand(100)
c2=corr(x2,y2)
data = 10 * np.random.rand(100) 
x3 = np.sort(data)              
epsilon3 = np.random.rand(100)  
y3 = (x3-5)**2 / 10 + epsilon3 - 0.5
c3=corr(x3,y3)
x4=[]
y4=[]
with open("mutant.txt","r") as inp:
    for line in inp:
        if line[0]=="C":
            continue
        x4.append(float(line.split()[0]))
        y4.append(float(line.split()[1]))
c4=corr(x4,y4)
x5=[]
y5=[]
with open("wild_type.txt","r") as inp:
    for line in inp:
        if line[0]=="C":
            continue
        x5.append(float(line.split()[0]))
        y5.append(float(line.split()[1]))
c5=corr(x5,y5)
print("Коэффицент корреляции")
print("Для набора данных n1:",c1)
print("Для набора данных n2, это полностью случайные данные:",c2)
print("Для набора данных n3, случайные данные которые отсортировали:",c3)
print("Для набора данных n4, эксперимент с мутантом:",c4)
print("Для набора данных n5, экспиримент с обычными клетками:",c5)
```
