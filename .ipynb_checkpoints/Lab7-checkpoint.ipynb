Упражнение 2

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
students = pd.read_csv("StudentsPerformance.csv")
bach=students[students["parental level of education"]=="bachelor's degree"]
col=students[students["parental level of education"]=="some college"]
mas=students[students["parental level of education"]=="master's degree"]
ass=students[students["parental level of education"]=="associate's degree"]
highs=students[students["parental level of education"]=="high school"]
shigh=students[students["parental level of education"]=="some high school"]
degree=["bachelor's degree", "some college", "master's degree", "associate's degree", "high school", "some high school"]
sp=plt.subplot(311)
plt.boxplot((bach["reading score"],col["reading score"],mas["reading score"],ass["reading score"],highs["reading score"],shigh["reading score"]),labels=degree)
plt.xlabel('Parental degree')
plt.ylabel('Reading score')
sp=plt.subplot(312)
plt.boxplot((bach["writing score"],col["writing score"],mas["writing score"],ass["writing score"],highs["writing score"],shigh["writing score"]),labels=degree)
plt.xlabel('Parental degree')
plt.ylabel('Writing score')
sp=plt.subplot(313)
plt.boxplot((bach["math score"],col["math score"],mas["math score"],ass["math score"],highs["math score"],shigh["math score"]),labels=degree)
plt.xlabel('Parental degree')
plt.ylabel('Math score')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("miceprotein.csv",na_values='?')
plt.figure(figsize=(15,10))
df=df.dropna(subset=["DYRK1A_N","SOD1_N"])
classes=["c-CS-m", "c-SC-m", "c-CS-s", "c-SC-s", "t-CS-m", "t-SC-m", "t-CS-s", "t-SC-s"]
cCSm=df[df['class']=="'c-CS-m'"]
cSCm=df[df['class']=="'c-SC-m'"]
cCSs=df[df['class']=="'c-CS-s'"]
cSCs=df[df['class']=="'c-SC-s'"]
tCSm=df[df['class']=="'t-CS-m'"]
tSCm=df[df['class']=="'t-SC-m'"]
tCSs=df[df['class']=="'t-CS-s'"]
tSCs=df[df['class']=="'t-SC-s'"]
sp=plt.subplot(211)
plt.boxplot((cCSm["DYRK1A_N"], cSCm["DYRK1A_N"], cCSs["DYRK1A_N"], cSCs["DYRK1A_N"], tCSm["DYRK1A_N"], tSCm["DYRK1A_N"], tCSs["DYRK1A_N"], tSCs["DYRK1A_N"]),labels=classes)
plt.xlabel('Mice class')
plt.ylabel('Expression level')
sp=plt.subplot(212)
plt.boxplot((cCSm["SOD1_N"], cSCm["SOD1_N"], cCSs["SOD1_N"], cSCs["SOD1_N"], tCSm["SOD1_N"], tSCm["SOD1_N"], tCSs["SOD1_N"], tSCs["SOD1_N"]),labels=classes)
plt.xlabel('Mice class')
plt.ylabel('Expression level')
plt.show()
```
